```python
def cntpairswithdiffgtk(sortednums, k):
  cntpairs = 0
  last = 0
  ln = len(sortednums)
  for first in range(ln):
    while last < ln and sortednums[last] - sortednums[first] <=k:
      last+=1
    cntpairs += ln - last
  return cntpairs


n, r = map(int,input().split())
A = list(map(int,input().split()))
print(cntpairswithdiffgtk(A, r))
```

```python
def makeprefixsum(nums):
  prefixsum = [0]*(len(nums)+1)
  for i in range(1, len(nums)+1):
    prefixsum[i] = prefixsum[i-1] + nums[i-1]
  return prefixsum

def rsq(prefixsum, l, r):
  return prefixsum[r] - prefixsum[l]

def sumtripletimes(nums):
  ans = 0
  sm = 0
  first = 0
  prefixsum = makeprefixsum(nums)
  ln = len(nums)
  for j in range(1,ln - 1):
    sm += nums[j]*rsq(prefixsum, j+1, ln)

  for first in range(ln- 2):
    ans += nums[first]*sm
    sm -= nums[first+1]*rsq(prefixsum, first+2, ln)

  return ans % 1000000007

n = int(input())
A = list(map(int, input().split()))
print(sumtripletimes(A))
```

```python
import random

def generate_random_lists(num_lists=5000, max_length=1000, max_value=1000):
    random_lists = []
    for _ in range(num_lists):
        # Определяем случайную длину списка от 1 до max_length
        length = random.randint(1, max_length)
        # Генерируем случайный список указанной длины
        random_list = [random.randint(1, max_value) for _ in range(length)]
        random_lists.append(random_list)
    return random_lists



def makeprefixsum(nums, k):
  prefixsum = [[1, k] for _ in range(len(nums))]
  if k != 0:
    for i in range(1, len(nums)):
      if nums[i] == nums[i - 1]:
        if prefixsum[i - 1][1] > 0:
          prefixsum[i] = [prefixsum[i - 1][0], prefixsum[i - 1][1] - 1]
        elif prefixsum [i - 1][1] == 0:
          prefixsum[i] = [prefixsum [i - 1][0] + 1, 0]
      elif nums[i] > nums[i - 1]:
        prefixsum[i] = prefixsum[i - 1]
      else:
        prefixsum[i] = [i + 1, k]
  if k == 0:
    for i in range(1, len(nums)):
      if nums[i] == nums[i - 1]:
        prefixsum[i] = [i + 1, 0]
      elif nums[i] > nums[i - 1]:
        prefixsum[i] = prefixsum[i - 1]
      else:
        prefixsum[i] = [i + 1, k]
  return prefixsum

def fast(n, evidences, m, k, numbers):
  prefixsum = makeprefixsum(evidences, k)
  ans = []
  for i in range(m):
    ans.append(prefixsum[numbers[i] - 1][0])
  return ans



def slow(n, evidences, m, k, numbers): #работает правильно!
  ans = []
  for i in range(m):
    patience = 0
    pointer = numbers[i] - 1
    while pointer > 0:
      if evidences[pointer - 1] <= evidences[pointer]:
        if evidences[pointer - 1] == evidences[pointer]:
          patience += 1
          if patience > k:
            break
        pointer -= 1
      else:
        break
    ans.append(pointer + 1)
  return ans


A = generate_random_lists()

for B in A:

  n = len(B)
  k = random.randint(0, n)
  numbers = [i + 1 for i in range(n)]
  m = len(numbers)
  flag = 1
  if slow(n, B, m, k, numbers) != fast(n, B, m, k, numbers):
    print('test error', B, k)
    flag = 0

if flag == 1:
  print('all tests are correct')
```
