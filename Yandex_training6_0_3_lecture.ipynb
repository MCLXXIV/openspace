Конструкция for-else в Python работает следующим образом: блок else выполняется после завершения цикла for, если цикл не был прерван оператором break. Это бывает полезно, если нужно отличить нормальное завершение цикла от ситуации, когда из цикла вышли досрочно.

```python
#ПСП
s = input()
d = {'(': ')', '[': ']', '{': '}'}
stack = []
for i in s:
    if i in d:
        stack.append(i)
    elif stack and d[stack[-1]] == i:
        stack.pop()
    else:
        print('no')
        break
else:
    print('yes' if not stack else 'no')
```

```python
#первый меньше справо

n = int(input())
A = list(map(int, input().split()))
ans = [-1]*n
stack = []
for i in range(n):
  if stack:
    j = len(stack) - 1
    while stack and stack[j][0] > A[i]:
      ans[stack[j][1]] = i
      stack.pop()
      j-=1
  stack.append([A[i], i])


print(*ans)
```

```python
n = int(input())
A = list(map(int, input().split()))
ans = [-1] * n
stack = []

for i in range(n):
    # Удаляем из стека все элементы, которые больше текущего элемента A[i]
    while stack and stack[-1][0] > A[i]:
        _, index = stack.pop()
        ans[index] = i
    # Добавляем текущий элемент и его индекс в стек
    stack.append((A[i], i))

print(*ans)
```

```python
#минимум в окне

n, k = map(int, input().split())
A = list(map(int, input().split()))

ans = []
stack = []
for i in range(k):
  if stack:
    while stack and stack[-1] > A[i]:
      stack.pop()
  stack.append(A[i])
ans.append(stack[0])
for i in range(n - k):
  if A[i] == stack[0]:
    stack.pop(0)
  while stack and stack[-1] > A[i + k]:
    stack.pop()
  stack.append(A[i + k])
  ans.append(stack[0])

print(*ans, sep = '\n')

```

```python
from collections import deque

n, k = map(int, input().split())
A = list(map(int, input().split()))

ans = []
deque_window = deque()
for i in range(k):
    while deque_window and A[deque_window[-1]] > A[i]:
        deque_window.pop()
    deque_window.append(i)
ans.append(A[deque_window[0]])

for i in range(k, n):
    if deque_window[0] == i - k:
        deque_window.popleft()
    while deque_window and A[deque_window[-1]] > A[i]:
        deque_window.pop()
    deque_window.append(i)
    ans.append(A[deque_window[0]])

print(*ans, sep='\n')
```

```python
#арифмитическое выражение

def split_multiple_delimiters(s, delimiters):
  result = []
  current = ''


  for char in s:
    if char in delimiters:
      if char == '-' and len(current.split()) == 0 and len(result) == 0:
        result.append('0')
      else:
        if char == '-' and len(current.split()) == 0 and result[-1] == '(':
           result.append('0')
        elif current:
          if len(current.split()) == 1:
            result.append(current.split()[0])
          elif len(current.split()) > 1 :
            result.append(current)
          current = ''
      result.append(char)

    else:
      current += char

  if current:
    if len(current.split()) == 1:
      result.append(current.split()[0])
    else:
      result.append(current)

  return result


def postfix(s):
  try:
    s = s.split()
    stack = []
    for i in s:
        if i == '+':
            b = stack.pop()
            a = stack.pop()
            stack.append(a + b)
        elif i == '*':
            b = stack.pop()
            a = stack.pop()
            stack.append(a * b)
        elif i == '-':
            b = stack.pop()
            a = stack.pop()
            stack.append(a - b)
        else:
            stack.append(int(i))  # Число добавляется как int
    return stack[0]
  except:
    return "WRONG"


A = input()
stack = []
ans = []
priority = {'+' : 0, '-' : 0, '*': 1}
delimetrs = '+-*()'
A = split_multiple_delimiters(A, delimetrs)
flag = True
for i in range(len(A)):
  if A[i] == '(':
    if (i == len(A) - 1) or (A[i+1] != '(' and not A[i + 1].isdigit()):
      flag = False
      break
else:
  for i in range(len(A)):
    if A[i].isdigit():
      ans.append(A[i])
    elif A[i] == '(':
      stack.append('(')
    elif A[i] == ')':
      while stack and stack[-1] != '(':
        ans.append(stack[-1])
        stack.pop()
      if not stack:
        flag = False
        break
      stack.pop()
    else:
      if A[i] in priority:
        while stack and stack[-1] != '(' and priority[stack[-1]] >= priority[A[i]]:
          ans.append(stack[-1])
          stack.pop()
        stack.append(A[i])
      else:
        flag = False
        break
if flag and not stack.count('(') and not stack.count(')'):
  while stack:
    ans.append(stack[-1])
    stack.pop()
  print(postfix(' '.join(ans)))
else:
  print('WRONG')

#точно можно проще


#не должно быть пробелов между числами - это точно
#не должно быть не понятных символов - это точно
#не должно наверное быть неправилльных скобок?
#c минус корректно должно работать! добавлять 0 после скобок и в начале выражения
```

```python
#минимальная псп
from collections import deque

d = deque()

n = int(input())
priority = input()
s = input()

d1 = {'(': ')', '[': ']'}

stack = []
for i in s:
    if i in d1:
        stack.append(i)
    elif stack and d1[stack[-1]] == i:
        stack.pop()

for i in stack:
  d.appendleft(d1[i])

counter = n - len(s) - len(d)

for i in range(n - len(s)):
  for j in priority:
    if d and j == d[0]:
      s += j
      d.popleft()
      break
    elif j in d1 and counter:
      s += j
      d.appendleft(d1[j])
      counter -= 2
      break

print(s)
#([]([][]))
```
